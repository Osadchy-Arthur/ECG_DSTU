```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from tensorflow.keras import layers, losses
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
```

```python
df = pd.read_csv('ecg.csv', header=None)
df.head()
```

```python
#Теперь мы разделим данные и метки так, чтобы нам было легко это сделать

#iloc используется для выбора данных по индексам (integer-location based indexing).
data = df.iloc[:,:-1].values
labels = df.iloc[:,-1].values
labels
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size = 0.2, random_state = 21)
print(train_data)
```

**Позволяет нормализовать данные к диапазону [0-1]**

Используем простую формулу:-


```python
#Теперь давайте нормализуем данные
#Сначала мы вычислим максимальное и минимальное значения из обучающего набора
min = tf.reduce_min(train_data)
max = tf.reduce_max(train_data)

#Теперь мы будем использовать формулу (data - min)/(max - min)
train_data = (train_data - min)/(max - min)
test_data = (test_data - min)/(max - min)

#преобразуем данные в float
train_data = tf.cast(train_data, dtype=tf.float32)
test_data = tf.cast(test_data, dtype=tf.float32)
```

```python
#Метки равны либо 0, либо 1, поэтому преобразуем их в логическое значение(true или false).
train_labels = train_labels.astype(bool)
test_labels = test_labels.astype(bool)

#Теперь давайте отделим данные нормальной ЭКГ от данных аномальной ЭКГ
#Данные нормальной ЭКГ
n_train_data = train_data[train_labels]
n_test_data = test_data[test_labels]

#Аномальные данные ЭКГ
an_train_data = train_data[~train_labels]
an_test_data = test_data[~test_labels]

print(n_train_data)
```

```python
#Давайте построим нормальную ЭКГ
plt.plot(np.arange(140), n_train_data[0])
plt.grid()
plt.title('Normal ECG')
plt.show()
```

```python
#Давайте построим график на основе аномальной ЭКГ
plt.plot(np.arange(140), an_train_data[0])
plt.grid()
plt.title('Abnormal ECG')
plt.show()
```

```python
# Теперь давайте определим модель!
# Здесь я использовал API подклассов Model (но мы также можем использовать последовательный API)
# Модель состоит из 2 частей: 1. Кодировщик и 2. декодировщик
 
class detector(Model):
  def __init__(self):
    super(detector, self).__init__()
    self.encoder = tf.keras.Sequential([
                                        layers.Dense(32, activation='relu'),
                                        layers.Dense(16, activation='relu'),
                                        layers.Dense(8, activation='relu')
    ])
    self.decoder = tf.keras.Sequential([
                                        layers.Dense(16, activation='relu'),
                                        layers.Dense(32, activation='relu'),
                                        layers.Dense(140, activation='sigmoid')
    ])

  def call(self, x):
    encoded = self.encoder(x)
    decoded = self.decoder(encoded)
    return decoded
```

```python
#Компиляция и обучение модели
autoencoder = detector()
autoencoder.compile(optimizer='adam', loss='mae')
history = autoencoder.fit(n_train_data, n_train_data, epochs = 20, batch_size=512, validation_data=(n_test_data, n_test_data))
```

```python
#Теперь давайте определим функцию, которая будет отображать исходную ЭКГ и реконструированную, а также показывать ошибку
def plot(data, n):
  enc_img = autoencoder.encoder(data)
  dec_img = autoencoder.decoder(enc_img)
  plt.plot(data[n], 'b')
  plt.plot(dec_img[n], 'r')
  plt.fill_between(np.arange(140), data[n], dec_img[n], color = 'lightcoral')
  plt.legend(labels=['Input', 'Reconstruction', 'Error'])
  plt.show()

plot(n_test_data, 0)
plot(an_test_data, 0)
```

```python
reconstructed = autoencoder(n_train_data)
train_loss = losses.mae(reconstructed, n_train_data)
t = np.mean(train_loss) + np.std(train_loss)

def prediction(model, data, threshold):
  rec = model(data)
  loss = losses.mae(rec, data)
  return tf.math.less(loss, threshold)
print(t)
```

```python
pred = prediction(autoencoder, n_test_data, t)
print(pred)
```

```python

plot(n_test_data, 0)
plot(n_test_data, 1)
plot(an_test_data, 3)

def plot_loss(history):
    plt.figure(figsize=(10, 5))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

plot_loss(history)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Функция для предсказания аномалий
def prediction(model, data, threshold):
    rec = model(data)
    loss = losses.mae(rec, data)
    return tf.math.less(loss, threshold)

# Получение порога для определения аномалий
reconstructed = autoencoder(n_train_data)
train_loss = losses.mae(reconstructed, n_train_data)
threshold = np.mean(train_loss) + np.std(train_loss)

# Предсказание аномалий на тестовом наборе
pred_test = prediction(autoencoder, test_data, threshold)

# Преобразование предсказаний в numpy массив и приведение типов
pred_test = np.array(pred_test)
pred_test = pred_test.astype(int)

# Построение матрицы ошибок
cm = confusion_matrix(test_labels, pred_test)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])

# Визуализация матрицы ошибок
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```
